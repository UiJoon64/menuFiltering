```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import requests

who_list=[]
# url 주소
url = 'https://sauce.foodpolis.kr/home/index.do'

# 드라이버 열기
driver = webdriver.Chrome()
driver.get(url)

# 소스 전문 정보 버튼 클릭
element = driver.find_element(By.XPATH, "//*[@id='gnb']/ul/li[3]")
element.click()
for _ in range(1,83,1):
    for k in range(4,8,1):
        for i in range(1,13,1):
            element = driver.find_element(By.XPATH,'//*[@class="conTable tdCenter"]/tbody/tr[{}]'.format(i))
            who_list.append(element.text)
        # 페이지 버튼 클릭
        element = driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div[4]/ul/li[{}]'.format(k))
        element.click()
    for i in range(1,13,1):
        element = driver.find_element(By.XPATH,'//*[@class="conTable tdCenter"]/tbody/tr[{}]'.format(i))
        who_list.append(element.text)

    ### 넥스트 버튼 누르기
    element = driver.find_element(By.XPATH,'//*[@class="next"]/a')
    element.click()
### 마지막에서 2번째 페이지 크롤링
for i in range(1,13,1):
    element = driver.find_element(By.XPATH,'//*[@class="conTable tdCenter"]/tbody/tr[{}]'.format(i))
    who_list.append(element.text)
element = driver.find_element(By.XPATH,'//*[@id="content"]/div[2]/div[4]/ul/li[4]')
element.click()
### 마지막 페이지 크롤링
for i in range(1,2,1):
    element = driver.find_element(By.XPATH,'//*[@class="conTable tdCenter"]/tbody/tr[{}]'.format(i))
    who_list.append(element.text)

driver.quit()
    
who_list
```
