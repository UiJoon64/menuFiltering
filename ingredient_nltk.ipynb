```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import platform
import kss
from konlpy.tag import Okt

### 파일 불러오기
df = pd.read_csv("D:/pukyung_202301/miniproject/new_food.csv")

### 타입 문자열로 모두 변경
df["ingredient"] = df["ingredient"].astype(str)
in_df = df["ingredient"]

### 만약 문자열에 ]]]이 포함되지 않는다면 ]]]를 추가해줌.
#   (크롤링 시 누락된 것도 있어서)
pro_ingredient_list=[]
for ingredient in in_df:
    if "]]]" not in ingredient:
        if len(ingredient) < 5:
            pro_ingredient_list.append("[[[]]]")
        else:
            #print("수정 전 : \n", ingredient,"\n")
            processed_ingredient = ingredient+"']]]"
            #print("수정 후 : \n", processed_ingredient)
            pro_ingredient_list.append(processed_ingredient)
    elif len(ingredient) < 5:
        pro_ingredient_list.append("[[[]]]")
    else:
        pro_ingredient_list.append(ingredient)
        
### 3차원 -> 1차원으로 ""붙여주기
def flatten_list(lst):
    flat_list = []
    for sublist in lst:
        if isinstance(sublist, list):
            flat_list.extend(flatten_list(sublist))
        else:
            flat_list.append(sublist)
    return flat_list

### 폰트 설정
font_path = "c:/Windows/Fonts/malgun.ttf"

### 0~399 중 첫 메뉴 가져오기 ---------------------------------------------------
for idx,i in enumerate(in_df) :
    fin_list=[]
    #print(idx,'\n\n\n')
    first_df = in_df[idx].split("]")  # > []안에 숫자로 메뉴 바꾸기
   #print(len(first_df),"\n\n")
    for j in range(len(first_df)) :
        before_df = first_df[j]  # > []안에 숫자로 메뉴 속 레시피번호 바꾸기

        ### 토큰화(한글버전)
        df_token = kss.split_sentences(before_df)

        ### [] 없애기
        clean_df = before_df.replace("[", "")

        ### 단어 정제 ~
        okt = Okt()
        okt.pos(clean_df)

        verb_okt = okt.nouns(clean_df)
        fin_list.append(verb_okt)
        print(verb_okt)
    df["filtered_ingredient"][idx]=fin_list
    
    
# 이후 단어 정규화도 진행 후 파일에 append
```
