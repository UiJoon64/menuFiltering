```python
import numpy as np
import pandas as pd

# 데이터 읽어들이는 부분
df=pd.read_csv("./new_food.csv")
ingredients = df["final_ingredient"]
menu = df["음 식 명"]
wo2_list=[]
for iters in range(0,400,1):
    ingredients1 = ingredients[iters]
    menu1 = menu[iters]

    # 문자열을 리스트로 변환하는 과정, 변환하는 더 쉬운 과정있었음!
    ing_list1 = ingredients1.split("], [")
    ing_ori_list=[]

    for sentence in ing_list1:
        ing_ori_list.append(sentence.split(","))
    word_list=[]
    for word in ing_ori_list:
        word_list.append(word)

    # key에서 한글만 정제. 정규표현식 활용하면 코드 훨씬 더 줄일 수 있을 것
    list1=[]
    for lists in word_list:
        list2=[]
        for word in lists:
            word = word.replace("'","")
            word = word.replace("[","")
            word = word.replace("]","")
            list2.append(word.replace(" ",""))
        list1.append(list2)

    # dictionary의 각 key에 기본 0값을 추가
    diction={}
    for lists in list1:
        for word in lists:
            diction[word]=0

    # 중복돼서 나온 key들의 value에 값들을 하나씩 추가
    for lists in list1:
        for word in lists:
            diction[word]=diction[word]+1

    # 확률을 딕셔너리의 value값에 저장
    prob_dict={}
    for key,value in diction.items():
            prob_dict[key]=np.round(value/len(list1),2)
            
    prob_dict = dict(sorted(prob_dict.items(),key=lambda x:x[1],reverse=True))
    print(menu1,"\n\n",prob_dict,"\n\n",prob_dict.items(),"\n\n")
    wo_list=[]
    for wo in prob_dict.items():
        wo_list.append(wo[0])

        #여기 wo_list[:n] 부분이 top 몇 재료를 뽑을지 정하는 것

    wo2_list.append(wo_list[:5])
print(wo2_list)    
    #menu2.append(menu1)
df["top7_ingredients"]=wo2_list
```

```python
import re

final_ingred = []
for idx in range(0,400,1):
    ingred = df["final_ingredient"][idx].split("], [")[:-3]
    new_ingred = []
    for i in ingred:
        pattern = re.compile('[^ ㄱ-ㅣ가-힣]+')
        new_ingred.append(pattern.sub('',i))
    new_ingred
    new_new_ingred=[]
    for j in new_ingred:
        k = j.split(" ")
        new_new_ingred.append(k)
    final_ingred.append(new_new_ingred)
final_ingred
```

```python
list3=[]
for idx,i in enumerate(final_ingred):
    list2=[]
    for j in i:
        #print(idx,j)
        list1=[]
        for ingre in j:
            if ingre in df["top7_ingredients"][idx]:
                list1.append(True)
                #print(ingre)
        list2.append(list1)
    list3.append(list2)
list3
```

```python
final_review = []
for idx in range(0,400,1):
    review = df["review"][idx].split("], [")
    new_review = []
    for i in review:
        pattern2 = re.compile('[^ ㄱ-ㅣ가-힣]+')
        new_review.append(pattern2.sub('',i))

    new_new_review=[]
    for j in new_review:
        k = j.split(" ")
        new_new_review.append(k)
    final_review.append(new_new_review)
```

```python
for idx,i in enumerate(list3):
    appending_ingred=[]
    appending_review=[]
    for idx2,j in enumerate(i):
        try:
            if len(j)!=0:
                #print(idx,idx2)
                appending_ingred.append(final_ingred[idx][idx2])
                appending_review.append(final_review[idx][idx2])
        except:
            print(idx,idx2)
    df["real_ingred"][idx]=appending_ingred
    df["real_review"][idx]=appending_review
```
